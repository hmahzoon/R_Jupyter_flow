```r
#Building Decision Tree
#For rpart() we need to load rpart library
#install.packages("rpart")
library(rpart)

colnames(projectFinal)

#final_model <- rpart(Is_finalic ~ .,data = final_train)
final_model <- rpart(leftemployees ~ .,data = final_train)
final_model

plot(final_model,margin = 0.1)
text(final_model,use.n = TRUE,pretty = TRUE,cex =0.8)

#create subset and verify
temp<-final_train[final_train$satisfaction_level >=0.465 & final_train$time_spend_company < 4.5,]
table(temp$leftemployees)

#Prediction of test dataset

#pred_final<-predict(final_model,newdata = final_test,type = "class")
pred_final<-predict(final_model,newdata = final_test,type="class")
head(pred_final)

#Now we need to compare it with actual values
table(pred_final,final_test$leftemployees)
length(pred_final)
length(final_test$leftemployees)
#For creating confusion matrix we can use the following
library(caret)
confusionMatrix(table(pred_final,final_test$leftemployees))

#Continue from here.

```

```r
#Random Forest
library(randomForest)
final_forest<-randomForest(leftemployees ~ .,data = final_train, ntree = 10)

final_forest

#Prediction of Test set

pred1_final<-predict(final_forest,newdata = final_test,type = "class")
pred1_final
```

```r
#confusion matrix
library(caret)
confusionMatrix(table(pred1_final,final_test$leftemployees))

```

```r
# Naive Bayes Classifier

library(e1071)

final_naive<-naiveBayes(leftemployees ~ .,data = final_train)
final_naive

#Prediction of Test set

pred2_final <- predict(final_naive,newdata = final_test,type = "class")
pred2_final

#Confusion matrix
library(caret)
confusionMatrix(table(pred2_final,final_test$leftemployees))



```

```r


#Building svm model

model<- svm(leftemployees ~ ., data = final_train,kernel = "polynomial",cost = .01,scale = TRUE)
#model<- svm(leftemployees ~ ., data = final_train,kernel = "radial",cost = .1,scale = TRUE)

#plot(model,final_train)
model

#predicting values of test set
pred<-predict(model,final_test, type = "class")
pred

table(pred,final_test$leftemployees)
mean(pred==final_test$leftemployees)
library(ggplot2)
library(caret)
confusionMatrix(table(pred,final_test$leftemployees))

summary(model)
model$index

#model<-svm(leftemployees ~ .,data = final_train,kernel = "polynomial",cost = .1,scale = TRUE)
#plot(model,final_train)

#cross validation using tune
############IT TAKES A COUPLE OF MINUTES< BE PATIENT!##########
#tune.out<-tune(svm,leftemployees~.,data = final_train,kernel = "radial",ranges = list(cost = c(0.01,.1,1,10)))

#Polynomial works better!!! less time than radial!
tune.out<-tune(svm,leftemployees~.,data = final_train,kernel = "polynomial",ranges = list(cost = c(0.01,.1,1,10)))
#tune.out<-tune(svm,leftemployees~.,data = final_train,kernel = "polynomial",ranges = list(cost = c(.0001,.001,.01)))
#tune.out<-tune(svm,leftemployees~.,data = final_train,kernel = "polynomial",ranges = list(cost = c(1,50,100)))

summary(tune.out)

bm<-tune.out$best.model
summary(bm)

pred<-predict(bm,final_test, type = "class")

library(caret)
confusionMatrix(table(pred,final_test$leftemployees))

```
